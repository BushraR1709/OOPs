Q1. What is Abstraction in OOps? Explain with an example.

```python
# In OOPs with Python, Abstraction refers to the process of creating abstract classes and methods that define a common interface for a group of 
#related objects.
# It allows us to define the structure and behavior of objects without specifying the implementation details.
# Example:

import  abc
class Details:
    
    @abc.abstractmethod
    def personal(self):
        pass
    
    @abc.abstractmethod
    def academic(self):
        pass
    @abc.abstractmethod
    def experience(self):
        pass
    
class Student1(Details):
    def personal(self):
        return "Name of student is Bushra"
        
class Student2(Details):
    def personal(self):
        return "Name of student is Nisha"
    def academic(self):
        return 'Student is pursuing Data Science Masters course from PWskills'
    def experience(self):
        return 'Freshers'
    
```

```python
s1 = Student1()
```

```python
s2 = Student2()
```

```python
s1.personal()
```

```python
s2.personal()
```

```python
s2.academic()
```

```python
s2.experience()
```

Q2. Differentiate between Abstraction and Encapsulation. Explain with an example.

```python
# Abstraction refres to creating abstract classes and methods that define a common interface for a group of related objects. It helps in organizing and structuring 
# our codebase effectively.

# Encapsulation, on the other hand, is the practice of bundling data and methods together within a class. It allows for data hiding and provides a way to control access to 
# the internal state of an object.

#Example 
# 1 - Abstraction 

import abc
class Animal:
    
    @abc.abstractmethod
    def makeSound(self):
        pass
    
class Cat(Animal):
    def makeSound(self):
        return 'Mew-Mew'

class Dog(Animal):
    def makeSound(self):
        return 'Bow-Bow'
        
```

```python
C = Cat()
D = Dog()
```

```python
C.makeSound()
```

D.makeSound()

```python
#Example 2 - Encapsulation
class BankAccount:
     def __init__(self,Ac_num,Balance):
        self.__Ac_num = Ac_num
        self.__Balance = Balance
        
     def get_balance(self):
        return self.__Balance
        
```

Q4. How can we achieve data abstraction?

```python
#Data abstraction is a specific type of abstraction that focuses on hiding the implementation details of data structures 
#and exposing only the necessary operations or behaviors related to that data.
# To achieve data abstraction, we can use classes and define abstract methods.

from abc import ABC,abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
    @abstractmethod
    def perimeter(self):
        pass
    
class Square(Shape):
    def __init__(self,s):
        self.side = s
    def area (self):
        return self.side*self.side    
    def perimeter(self):
        return 4 *self.side
```

```python
Sh = Shape()
```

```python
# Hence, it is shown that we can't create object of abstract class 
```

```python
sq = Square(5)
```

```python
sq.area()
```

```python
sq.perimeter()
```
