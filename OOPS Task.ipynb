Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.


```python
# Multiple inheritance that allows a class to inherit from two or more parent classes.

class one():
    def method1(self):
        print("This is class one")
    
class two():
    def method2(self):
        print("This is class two")
    
class three(one,two):
    def method3(self):
        print("This is class three")  
```

```python
TH = three()
```

```python
TH.method3()
```

```python
TH.method1()
```

```python
TH.method2()
```

Q5.What is method overriding in python? Write a python code to demonstrate method overriding

```python
# Method overriding in Python is a concept where a subclass provides its own implementation of a method that is already defined in its superclass.
# This allows the subclass to customize the behavior of the inherit method.
```

```python
class Flower():
    def flower_color(self):
        print ("Each flower has different color ")
```

```python
class Rose(Flower):
    def flower_color(self):
        print("Rose are red in color")
        
```

```python
class Lotus(Flower):
    def flower_color(self):
        print("Lotus are pink in color")
```

```python
#creating instances of each class
flo=Flower()
```

```python
rose = Rose()
```

```python
lotus=Lotus()
```

```python
#calling the overriden methods
flo.flower_color()
```

```python
rose.flower_color()
```

```python
lotus.flower_color()
```
