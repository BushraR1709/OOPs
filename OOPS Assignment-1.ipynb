Q3. Explain why the __init__() function is used. Give a suitable example.

```python
# init()method acts as a constructor.
# init() constructor assign all the values at the time of creating an object

#Example:-

class student():
    def __init__(self,name,age,gender):
        self.name = name
        self.age = age
        self.gender = gender
        
    def student_details(self):
        print("Name of the student is: ", self.name)
        print("Age of the student is: ", self.age)
        print("Gender of the student is: ", self.gender)
        
```

```python
s = student('Bushra', 22, 'Female')
s.student_details()
```

Q5. What is inheritance? Give an example for each type of inheritance.

```python
# Inheritance is one of the important concept of OOPs in Python. With inheritance one class can derive the properties of another class.
# It has Super class & Sub class or we can say Parent class & child class. The properties of super class is inherit by sub class.
```

```python
# Ex-1 Sinle inheritance
class Superclass :
    def method_super(self):
        return ' This is super class'
```

```python
class Subclass(superclass):
    def method_sub(self):
        return 'this is sub class'
```

```python
obj_sub = Subclass()
```

```python
obj_sub.method_super()
```

```python
obj_sub.method_sub()
```

```python
# Ex-2 Multi-level inheritance
class Level1 :
    def method1(self):
        return 'this is level1'
```

```python
class Level2(Level1):
    def method2(self):
        return ' this is level2'
        
```

```python
class Level3(Level2):
    def method3(self):
        return 'this is level3'
```

```python
l3 = Level3()
```

```python
l3.method1()
```

```python
l3.method2()
```

```python
l3.method3()
```

```python
# Ex - 3 Multiple inheritance
class Dad:
    def d_method(self):
        return 'This is Dad class'
```

```python
class Mom:
    def m_method(self):
        return 'This is Mom class'
```

```python
class Baby(Dad,Mom):
    def b_method(self):
        return 'This is Baby class'
```

```python
son = Baby()
```

```python
son.d_method()
```

```python
son.m_method()
```

```python
son.b_method()
```

```python
# Ex - 4 Hierarchial inheritance
class Base:
    def base_method(self):
        return 'This is base class'
```

```python
class Derived_A(Base):
    def method_a(self):
        return 'This is derived A'
```

```python
class Derived_B(Base):
    def method_b(self):
        return 'This is derived B'
```

```python
A = Derived_A()
```

```python
B = Derived_B()
```

```python
A.base_method()
```

```python
A.method_a()
```

```python
B.base_method()
```

```python
B.method_b()
```

```python
# Ex - 5 Hybrid inheritance
class Base:
    def method_base(self):
        return 'This is Base class'
```

```python
class Alpha(Base):
    def method_alpha(self):
        return 'This is Alpha class'
```

```python
class Class_A(Alpha):
    def method_a(self):
        return 'Class A'
class Class_B(Alpha):
    def method_b(self):
        return 'Class B'
class Class_C(Alpha):
    def method_c(self):
        return 'Class C'
```

```python
A = Class_A()
B = Class_B()
C = Class_C()
```

```python
A.method_base()
```

```python
A.method_alpha()
```

```python
A.method_a()
```

```python
B.method_base()
```

```python
B.method_alpha()
```

```python
B.method_b()
```

```python
C.method_base()
```

```python
C.method_alpha()
```

```python
C.method_c()
```
